## One Dimensional GD

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
def gd(f_x, x_k: float = 1.0, lr: float = .1, e: float = 10**-3) -> None:
    f_prime = sp.diff(f, x)
    iter = 0
    while(abs(f_prime.subs(x, x_k)) > e):
        iter += 1
        x_k = x_k - lr * (f_prime.subs(x, x_k))
        print(f"@ iter #{iter}: x = {x_k}")
        
    return None
```

```python
def f(x):
    return x**2

x_axis = np.linspace(-10, 10, 100)  # Generate 100 x-values from -10 to 10
y_axis = f(x_axis)  # Calculate the corresponding y-values

plt.plot(x_axis, y_axis)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x) = x^2')
plt.grid(True)
plt.ylim(min(y_axis), max(y_axis))
plt.show()
```

```python
x = sp.Symbol('x')
f = x**2
gd(f, lr=.2)
```

```python
def f(x):
    return ((x**2) * (x-1) * (x+1.2))

x_axis = np.linspace(-10, 10, 100)  # Generate 100 x-values from -10 to 10
y_axis = f(x_axis)  # Calculate the corresponding y-values

plt.plot(x_axis, y_axis)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x) = (x^2) * (x-1) * (x+1.2)')
plt.grid(True)
plt.ylim(min(y_axis), max(y_axis))
plt.show()
```
