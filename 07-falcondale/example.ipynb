```python
from falcondale import Project
from sklearn import datasets
```

```python
# datasets.load_breast_cancer(() returns a tuple
# X are feature; y is 0/1 labels
X, y = datasets.load_breast_cancer(return_X_y=True, as_frame=True)
df = X.copy()
df['label'] = y
```

```python
myproject = Project(df, target="label")
myproject.profile_dataset()
```

```python
myproject.preprocess()
myproject.show_features()
```

```python
myproject.feature_selection(max_cols=10)
```

```python
model = myproject.evaluate("qnn", test_size=0.1, layers=1)
model.print_report()
```

```python
model.list_metrics()
```

```python
model.metric("accuracy")
```

```python
model.metric("auc")
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

y_pred = model.predict(X)

print(accuracy_score(y, y_pred))
print(confusion_matrix(y, y_pred))
print(classification_report(y, y_pred))
```
