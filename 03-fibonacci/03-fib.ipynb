```python
import matplotlib.pyplot as plt
import polars as pl
from tqdm import trange
```

```python
def gen_fib_seq(bi_nth: int, a: int = 0, b: int = 1) -> dict:
    """
    bi_nth: bidirectional nth fib number
    a, b: seq start
    """
    results = []
    for pos in trange(bi_nth * 2):
        # appends
        results.append(a)
        
        # when even, the exp (pos % 2 == 0) returns 1 and 'a' value is negatified
        # when odd, the exp (pos % 2 != 0) returns 1 and 'a' value is kept the same
        
        # inserts back
        results.insert(0, ((pos % 2 == 0) * -1) * a + (pos % 2 != 0) * a)
        
        # update
        a, b = b, a + b
        
    results.pop(len(results) // 2)
    return results
```

```python
SEQ_LENGTH = 10
fib_seq = gen_fib_seq(SEQ_LENGTH)
```

```python
plt.plot(fib_seq)
```
